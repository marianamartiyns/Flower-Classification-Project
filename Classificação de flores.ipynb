## Implementação de um Modelo de Classificação de Flores usando CNN

Este código implementa um modelo de rede neural convolucional (CNN) para classificação de imagens de flores usando aprendizado profundo (deep learning).

1. **Preparação dos dados:**

   - O código começa listando o número de imagens em cada subdiretório dentro do diretório 'imagens/'.
   - Os dados são então carregados usando a função `image_dataset_from_directory` do TensorFlow, divididos em conjuntos de treinamento e validação.

2. **Visualização dos dados:**

   - Algumas imagens de treinamento são visualizadas para entender melhor o conjunto de dados.
   

3. **Data Augmentation:**

   - Uma série de operações de data augmentation, como flip horizontal, rotação e zoom, são aplicadas às imagens de treinamento para aumentar a diversidade do conjunto de dados e melhorar a generalização do modelo.

4. **Definição do modelo CNN:**

   - Um modelo sequencial é criado usando a API Keras.
   - A camada de data augmentation é seguida por camadas de pré-processamento, convolucionais, de max pooling, dropout e camadas totalmente conectadas.
   - O modelo é compilado com uma função de perda de entropia cruzada categórica esparsa e otimizador Adam.
   - O modelo é treinado usando os dados de treinamento e validação.

   - Uma função é definida para classificar novas imagens após o treinamento do modelo.
   - Isso é demonstrado usando uma imagem de teste ('amostra/flor2.jpg').

```python
# importando as bibliotecas
import os
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import load_img, ImageDataGenerator
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten
```

```python
# Contagem do número total de imagens na pasta 'Imagens/'
count = 0 
dirs = os.listdir('Imagens/')
for dir in dirs:
    files = list(os.listdir('Imagens/'+dir))
    print( dir +' tem: '+ str(len(files)) + ' imagens')
    count = count + len(files)
print( 'A pasta Imagens tem: '+ str(count) + ' imagens')
```

```python
# Diretório base das imagens e definição de tamanho e batch size
base_dir = 'Imagens/'
img_size = 180
batch = 32
```

## Treinamento e Validação

```python
# Carregamento dos dados de treinamento e validação
train_ds = tf.keras.utils.image_dataset_from_directory( base_dir,
                                                       seed = 123,
                                                       validation_split=0.2,
                                                       subset = 'training',
                                                       batch_size=batch,
                                                       image_size=(img_size,img_size))

val_ds = tf.keras.utils.image_dataset_from_directory( base_dir,
                                                       seed = 123,
                                                       validation_split=0.2,
                                                       subset = 'validation',
                                                       batch_size=batch,
                                                       image_size=(img_size,img_size))
```

```python
# Nomes das classes de flores
flower_names = train_ds.class_names
flower_names
```

```python
# Visualização de algumas imagens de treinamento
i = 0
plt.figure(figsize=(10,10))

for images, labels in train_ds.take(1):
    for i in range(9):
        plt.subplot(3,3, i+1)
        plt.imshow(images[i].numpy().astype('uint8'))
        plt.title(flower_names[labels[i]])
        plt.axis('off')
```

## Otimização de dados

```python
# Configuração de otimização de dados

AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size = AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size = AUTOTUNE)
```

```python
# Definição das operações de data augmentation
data_augmentation = Sequential([
    layers.RandomFlip("horizontal", input_shape = (img_size,img_size,3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1)
])
```

```python
# Visualização das imagens com data augmentation aplicada
i = 0
plt.figure(figsize=(10,10))

for images, labels in train_ds.take(1):
    for i in range(9):
        images = data_augmentation(images)
        plt.subplot(3,3, i+1)
        plt.imshow(images[0].numpy().astype('uint8'))
        plt.axis('off')
```

## Criando o modelo

```python
# Definição da arquitetura do modelo CNN
model = Sequential([
    data_augmentation,
    layers.Rescaling(1./255),
    Conv2D(16, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Conv2D(32, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Conv2D(64, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Dropout(0.2),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(5)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

## Treinamento do modelo

```python
history = model.fit(train_ds, epochs=15, validation_data=val_ds)
```

```python
model.summary()
```

## Função para classificar imagens

```python
def classify_images(image_path):
    input_image = tf.keras.utils.load_img(image_path, target_size=(180,180))
    input_image_array = tf.keras.utils.img_to_array(input_image)
    input_image_exp_dim = tf.expand_dims(input_image_array,0)

    predictions = model.predict(input_image_exp_dim)
    result = tf.nn.softmax(predictions[0])
    outcome = 'Essa imagem pertence ao grupo de ' + flower_names[np.argmax(result)] + ' com uma pontuação/precisão de '+ str(np.max(result)*100)
    return outcome
```

```python
classify_images('amostra/flor2.jpg') #teste
```
